```python
import gzip
import json
import pandas as pd
```

```python
reviews_df = pd.read_json('Digital_Music.json.gz', lines=True)
```

```python
reviews_df.head()
```

```python
average_rating_per_item = reviews_df.groupby('asin')['overall'].mean()
least_rated_item = average_rating_per_item.idxmin()
most_rated_item = average_rating_per_item.idxmax()
print("Least Rated Item")
reviews_df[reviews_df['asin'] == least_rated_item].head()
```

```python
print("Most Rated Item")
reviews_df[reviews_df['asin'] == most_rated_item].head()
```

```python
reviews_df['review_length'] = reviews_df['reviewText'].str.len()
longest_reviews_item = reviews_df.loc[reviews_df['review_length'].idxmax()]
print(longest_reviews_item)
```

```python
reviews_df['reviewTime'] = pd.to_datetime(reviews_df['unixReviewTime'], unit='s').dt.strftime('%m-%d-%Y')
```

```python
reviews_df.head()
```

```python
parquet_file_path = "Digital_Music.parquet"
reviews_df.to_parquet(parquet_file_path, index=False)
```
